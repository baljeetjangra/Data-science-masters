#### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
from scipy import stats

sample_mean = 50
sample_std = 5
n = 50
confidence_level = 0.95

standard_error = sample_std / (n ** 0.5)
interval = stats.norm.interval(confidence_level, loc=sample_mean, scale=standard_error)

print("The 95% confidence interval is:", interval)
```

#### Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import scipy.stats as stats
import numpy as np

observed = [24, 18, 21, 9, 10, 18]  # observed counts of each color
expected = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]  # expected proportions of each color

total_observed = sum(observed)  # total number of M&Ms
expected_counts = [expected[i] * total_observed for i in range(len(expected))]  # expected counts of each color

# perform chi-square goodness of fit test
chi2, p = stats.chisquare(observed, expected_counts)

# determine if the result is significant
alpha = 0.05
if p < alpha:
    print("The result is significant. The bag does not match the expected distribution.")
else:
    print("The result is not significant. The bag matches the expected distribution.")
    
print("Chi-square value:", chi2)
print("p-value:", p)
```

#### Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

Ans - 

```python
from statsmodels.stats.proportion import proportion_confint

n = 500
p = 60/500

lower, upper = proportion_confint(p*n, n, alpha=0.05)

print(f"The 95% confidence interval for the true proportion of individuals who smoke is ({lower:.4f}, {upper:.4f})")
```

#### Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

```python
import scipy.stats as stats

sample_mean = 75
sample_std = 12
sample_size = 100   # assuming a sample size of 100

z_score = stats.norm.ppf(0.95)  # finding the z-score for 90% confidence level

# calculating the margin of error
margin_of_error = z_score * (sample_std / (sample_size**0.5))

# calculating the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"90% confidence interval: ({lower_bound}, {upper_bound})")
```

#### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = 10  # degrees of freedom
x = np.linspace(0, 30, 1000)  # range of chi-square values to plot
y = stats.chi2.pdf(x, df)  # chi-square PDF values for each value of x

# Plot the chi-square distribution
plt.plot(x, y, label="df=10")

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 1000)
y_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color="blue", alpha=0.5)

# Add labels to the plot
plt.xlabel("Chi-Square Value")
plt.ylabel("Probability Density")
plt.title("Chi-Square Distribution with df=10")
plt.legend()

# Display the plot
plt.show()
```

#### Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

Ans - To calculate the 99% confidence interval for the true proportion of people in the population who prefer Coke, we can use the following formula:

CI = p ± z * sqrt((p*(1-p))/n)

where CI is the confidence interval, p is the sample proportion, z is the z-score corresponding to the desired confidence level, and n is the sample size.

In this case, we have p = 520/1000 = 0.52, n = 1000, and we need to find the z-score for a 99% confidence level. Using a standard normal distribution table or a calculator, we can find that the z-score corresponding to a 99% confidence level is approximately 2.576.

Plugging in the values, we get:

CI = 0.52 ± 2.576 * sqrt((0.52*(1-0.52))/1000)

Simplifying the expression, we get:

CI = 0.52 ± 0.049

Therefore, the 99% confidence interval for the true proportion of people in the population who prefer Coke is (0.471, 0.569). This means that we can be 99% confident that the true proportion of people in the population who prefer Coke falls within this range.

#### Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

Ans- To conduct a chi-square goodness of fit test, we need to calculate the expected frequencies for each category assuming the null hypothesis (the coin is fair). For a fair coin, we expect the probability of getting tails to be 0.5 and the probability of getting heads to be 0.5.

We can use these probabilities to calculate the expected frequencies for each category as follows:

Expected frequency of tails = 0.5 x 100 = 50
Expected frequency of heads = 0.5 x 100 = 50
We can then use these expected frequencies and the observed frequencies to calculate the chi-square statistic and the p-value.

```python
import scipy.stats as stats

observed_freq = [45, 55]  # tails, heads
expected_freq = [50, 50]  # assuming a fair coin

chi2, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)

print("Chi-square statistic:", chi2)
print("p-value:", p_value)
```

#### Q9. A study was conducted to determine if there is an association between smoking status (smoker or non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis.

#### index Lung Cancer: Yes Lung Cancer: No
#### Smoker 60 140
#### Non-smoker 30 170

Ans - To conduct a chi-square test for independence, we need to set up the hypotheses:

Null hypothesis: There is no association between smoking status and lung cancer diagnosis.
Alternative hypothesis: There is a significant association between smoking status and lung cancer diagnosis.
We can use a significance level of 0.05 for this test.

We can calculate the expected frequencies for each cell assuming the null hypothesis is true:

Lung Cancer: Yes Lung Cancer: No Total
Smoker 75 125 200
Non-smoker 15 185 200
Total 90 310 400

We can calculate the chi-square statistic using the formula:

chi2 = sum((observed - expected)**2 / expected)

where the sum is taken over all cells in the contingency table.

In Python, we can perform the chi-square test using the scipy.stats module:

```python
import scipy.stats as stats

observed = [[60, 140], [30, 170]]
chi2, p, dof, expected = stats.chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("p-value:", p)
```

#### U.K. (n=500) - 225 - 175 - 100

Ans - We can use a chi-square test for independence to determine if there is a significant association between chocolate preference and country of origin.

First, we need to set up the null and alternative hypotheses:

Null hypothesis: There is no significant association between chocolate preference and country of origin.
Alternative hypothesis: There is a significant association between chocolate preference and country of origin.
We can set the significance level to 0.05.

Next, we need to calculate the expected frequencies under the assumption of independence between chocolate preference and country of origin. We can do this by taking the row and column totals and using them to calculate the expected frequencies. For example, the expected frequency for U.S. milk chocolate preference can be calculated as:

(200+150+150) * (200+225) / 1000 = 180

We can repeat this for all the other cells to obtain the following expected frequencies:

Milk Chocolate Dark chocolate White chocolate

U.S. 180 135 185
U.K. 245 185 70

Now we can calculate the chi-square statistic using the formula:

X^2 = Σ (observed frequency - expected frequency)^2 / expected frequency

We obtain:

X^2 = (200-180)^2/180 + (150-135)^2/135 + (150-185)^2/185 + (225-245)^2/245 + (175-185)^2/185 + (100-70)^2/70
= 13.62

Using a chi-square distribution table with (3-1)*(2-1) = 2 degrees of freedom (df), we find that the critical value for a significance level of 0.05 is 3.84. Since our calculated chi-square statistic (13.62) is greater than the critical value (3.84), we reject the null hypothesis and conclude that there is a significant association between chocolate preference and country of origin.

Therefore, we can say that there is evidence to suggest that the proportion of people who prefer milk chocolate, dark chocolate, or white chocolate is different in the U.S. versus the U.K.

#### Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

Ans - We can use a t-test to conduct this hypothesis test.

The null hypothesis is that the population mean is equal to 70:


H0: μ = 70
The alternative hypothesis is that the population mean is not equal to 70:


Ha: μ ≠ 70
We can use a two-tailed t-test because we are testing for a difference in either direction from 70.

We calculate the t-statistic as:


t = (x̄ - μ) / (s / sqrt(n))
where x̄ is the sample mean, μ is the hypothesized population mean (70), s is the sample standard deviation, and n is the sample size.

We can then use the t-distribution with n-1 degrees of freedom to find the p-value. If the p-value is less than the significance level, we reject the null hypothesis.

```python
import scipy.stats as stats

# Sample statistics
x_bar = 72
s = 10
n = 30

# Null hypothesis
mu = 70

# Calculate t-statistic and p-value
t = (x_bar - mu) / (s / (n**0.5))
p_value = 2 * stats.t.cdf(-abs(t), n-1)

print("t-statistic:", t)
print("p-value:", p_value)
```
